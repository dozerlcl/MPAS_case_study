# 下载ERA5I的脚本测试部分

## 下载任务时间控制
使用pd.datetime.strftime 控制时间字符串格式

```python
time_3hr    = pd.date_range('1994-04-01T00:00', '1994-04-03T00:00', freq = '3h')
time_3hr_da = xr.DataArray(time_3hr, name = "time")
datestring  = time_3hr.strftime("%Y-%m-%dT%H:%M")
```

## 提交任务部分

# 补充MPAS中ungrib缺少文件

```python
time_6hr    = pd.date_range('1996-03-01T00:00', '1996-03-22T00:00', freq = '6h')
time_6hr_da = xr.DataArray(time_6hr, name = "time")
datestring  = time_6hr.strftime("%Y-%m-%dT%H:%M")
datestring
```

## SFC部分

```python
#!/usr/bin/env python

from ecmwfapi import ECMWFDataServer

for ind,single_datestring in enumerate(time_6hr[0:-1]):

    start_date = single_datestring.strftime("%Y-%m-%d")
    end_date   = time_6hr[ind+1].strftime("%Y-%m-%d")

    server = ECMWFDataServer()
    server.retrieve({
        "class": "ei",
        "dataset": "interim",
        "date": start_date + "/to/" + end_date,
        "expver": "1",
        "grid": "0.75/0.75",
        "levtype": "sfc",
        "param": "31.128/34.128/172.128",
        "step": "6",
        "stream": "oper",
        "time": str(single_datestring.strftime("%H:%M:%S")),
#         "time" : "06:00:00",
        "type": "an",
        "target": "./download_files/1996_shotdown/sfc/ERA5I_SFC_" + str(single_datestring.strftime("%Y-%m-%dT%H:%M"))+ ".grb",
    })
    
#     break
```
