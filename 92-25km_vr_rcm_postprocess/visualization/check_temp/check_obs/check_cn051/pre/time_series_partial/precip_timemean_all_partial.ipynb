```python
from timeseries_count import *

# get partial mask_res collect into a dict
partial_list = ["NE", "NC", "YZ", "SW"]
path_in = "/raid52/yycheng/MPAS/REFERENCE/TEMP_DATA/partial_pre"
file_ref= "ydaymean_mask_sel_CN05.1_Pre_1961_2018_daily_025x025.nc"
file_vr = "ydaymean_mask_vr_pre_98-17.nc"
file_rcm = "ydaymean_mask_rcm_pre_98-17.nc"
partial_mean = {} # partial mask_res
partial_std = {} # partial mask_res
for i in partial_list:
    path_in_partial = path_in + "/" + i + "/mask_res_dayofyear" # refine path
    print(path_in_partial)
    partial_mean[i] = mean_count(path_in_partial, file_ref, "pre", file_vr, "precip_MPAS", file_rcm, "precip_MPAS")
    partial_std[i] = std_count(path_in_partial, file_ref, "pre", file_vr, "precip_MPAS", file_rcm, "precip_MPAS")
```

```python
mean_temp = {}
std_temp  = {}
time_idx = pd.date_range(start = "2003-04-01", end = "2003-08-31",freq='d') # add time coords

for i in partial_list:
    temp_obs = partial_mean[i][0].to_dataframe().rename(columns={"pre":"cn051"})
    temp_vr  = partial_mean[i][1].to_dataframe().rename(columns={"precip_MPAS":"vr"})
    temp_rcm = partial_mean[i][2].to_dataframe().rename(columns={"precip_MPAS":"rcm"})

    std_obs = partial_std[i][0].to_dataframe().rename(columns={"pre":"cn051"})
    std_vr  = partial_std[i][1].to_dataframe().rename(columns={"precip_MPAS":"vr"})
    std_rcm = partial_std[i][2].to_dataframe().rename(columns={"precip_MPAS":"rcm"})

    # vr_bias_value = np.squeeze(temp_vr.values - temp_obs.values) # 去掉冗余的维度
    # temp_vr_bias  = pd.Series(vr_bias_value,index=temp_obs.index,name = 'VR - CN05.1')
    
    # rcm_bias_value = np.squeeze(temp_rcm.values - temp_obs.values) # 去掉冗余的维度
    # temp_rcm_bias  = pd.Series(rcm_bias_value,index=temp_obs.index,name = 'RCM - CN05.1')

    temp_rcm['region'] = i # new region type

    mean_temp[i] = pd.concat([temp_obs, temp_vr, temp_rcm], axis=1)
    # mean_temp[i] = pd.concat([temp_vr_bias, temp_rcm_bias], axis=1)
    mean_temp[i].index = time_idx

    std_temp[i] = pd.concat([std_obs, std_vr, std_rcm], axis=1)
```

```python
import matplotlib.dates as dates
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)

# import matplotlib.pyplot as plt
# plt.style.use(['science', 'no-latex', 'cjk-tc-font'])
plt.style.use(['science', 'no-latex']) # 采用 安装的scinence绘图风格，matplotlib风格拓展

inum = 0
fig,ax = plt.subplots(4,1,figsize=(16,12))
for i in partial_list:
    plot_pd = mean_temp[i]
    std_pd  = std_temp[i]
    # pd_plot = pd.concat([df_obs, df_mod,minus], axis=1)
    ymin = -1.
    ymax = 16.
    # idx = pd.date_range("2003-"+'04-01', "2003-"+'09-01')
    # s = pd.Series(np.random.randn(len(idx)), index=idx)

    # fig, ax = plt.subplots()
    # ax.plot_date(idx.to_pydatetime(), 'v-')
    ax[inum] = plot_pd.plot(figsize=(10,12),style=['-k','-r','-b'],ax = ax[inum])
    # ax.xaxis.set_minor_formatter(dates.DateFormatter('%d\n%a'))
    ax[inum].fill_between(plot_pd.index, plot_pd['vr'].values+std_pd['vr'].values, plot_pd['vr'].values-std_pd['vr'].values, color = 'r', alpha = 0.2)
    ax[inum].fill_between(plot_pd.index, plot_pd['rcm'].values+std_pd['rcm'].values, plot_pd['rcm'].values-std_pd['rcm'].values, color = 'b', alpha = 0.2)
    ax[inum].fill_between(plot_pd.index, plot_pd['cn051'].values+std_pd['cn051'].values, plot_pd['cn051'].values-std_pd['cn051'].values, color = 'k', alpha = 0.2)
    # ax[inum].set_xlim([xmin,xmax])
    ax[inum].xaxis.grid(True, which="minor", linestyle = ":")
    ax[inum].xaxis.grid(True, which="major", linestyle = "-")
    ax[inum].yaxis.grid(True, which="minor", linestyle = ":")
    ax[inum].yaxis.grid(True, which="major", linestyle = "-")

    ax[inum].yaxis.set_major_locator(MultipleLocator(5.))
    ax[inum].yaxis.set_major_formatter(FormatStrFormatter('%1.f'))
    ax[inum].yaxis.set_minor_locator(MultipleLocator(1.))

    ax[inum].set_ylim([ymin,ymax])
    ax[inum].set_ylabel(i)
    ax[inum].get_legend().remove()
    ax[inum].xaxis.set_major_locator(dates.MonthLocator())
    ax[inum].xaxis.set_major_formatter(dates.DateFormatter('%b'))
    ax[inum].xaxis.set_minor_locator(dates.DayLocator(interval=2))
    inum = inum + 1

ax[0].legend()
ax[0].set_title("1998-2017 daily precipitation")
# 背景设置为白色，否则会成为透明，png格式下
# fig.patch.set_facecolor('white')
plt.savefig("./output_pic/98-17_pre_series_dayofyear.png", dpi = 300)
```
