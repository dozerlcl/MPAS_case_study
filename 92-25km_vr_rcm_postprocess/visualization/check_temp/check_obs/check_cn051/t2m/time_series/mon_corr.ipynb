计算每个月的时间平均之后的空间pattern的相关系数

```python
# # rename coordinate name
# ds_cn051_mean['t2m'].dims
rename_coords = dict(zip(ds_cn051_mean['tm'].dims, ds_vr_mean['t2m'].dims))
ds_cn051_mean['tm'] = ds_cn051_mean['tm'].rename(rename_coords)

# ds_vr_mean['precip_MPAS'].resample(Time="1MS").mean()

vr_corr_temp = xr.corr(ds_cn051_mean['tm'].resample(Time="1MS").mean(), \
    ds_vr_mean['t2m'].resample(Time="1MS").mean() , dim=['latitude','longitude']).dropna(dim='Time')
vr_monmean = dict(vr_corr_temp.groupby(vr_corr_temp.Time.dt.year))

rcm_corr_temp = xr.corr(ds_cn051_mean['tm'].resample(Time="1MS").mean(), \
    ds_rcm_mean['t2m'].resample(Time="1MS").mean() , dim=['latitude','longitude']).dropna(dim='Time')
rcm_monmean = dict(rcm_corr_temp.groupby(rcm_corr_temp.Time.dt.year))
```
