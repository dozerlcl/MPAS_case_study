```python
# %matplotlib inline
import numpy as np
from scipy.fftpack import dct, idct, dctn, idctn
import matplotlib.pyplot as plt
import xarray as xr
import pandas as pd
```

```python
# 2D
def f(x, y, L):
    return np.sin(np.pi*5*np.sqrt(x ** 2 + y ** 2))

L = 100
x = np.linspace(-1, 1, L)
y = np.linspace(-1, 1, L)
X, Y = np.meshgrid(x, y)
Z = f(X, Y, L)

fig = plt.figure()
# ax = plt.axes(projection='3d')
plt.contourf(x, y, Z)
```

```python
 from scipy import fftpack

res_dctn = dctn(Z, type=2, axes=[0,1])

# Now shift the quadrants around so that low spatial frequencies are in
# the center of the 2D fourier transformed image.
ps_2d_shift = fftpack.fftshift( res_dctn )
# ps_2d_shift = res_dctn

# Calculate a 2D power spectrum
psd2D = np.abs( ps_2d_shift )**2

plt.contourf(psd2D)
plt.colorbar()
```
