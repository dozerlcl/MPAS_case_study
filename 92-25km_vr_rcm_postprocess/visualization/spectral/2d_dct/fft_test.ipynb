对于FFT的理解，练习使用

```python
import numpy as np
from scipy.fft import fft, fftfreq, ifft
# Number of sample points
N = 600
# sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N, endpoint=False)
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(100.0 * 2.0*np.pi*x)
yf = ifft(y)
xf = fftfreq(N, T)[:N//2]
# xf = fftfreq(N, T)
import matplotlib.pyplot as plt
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2])) # 使用一半
# plt.plot(xf, 2.0/N * np.abs(yf))
plt.grid()
plt.show()
```
