```python
# %matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import pandas as pd
import scipy.fftpack
```

```python
dir_in = "/raid52/yycheng/MPAS/VR_postprocess/VR_merge/ke_vi_daily/"
ds_vr_ke = xr.open_dataset(dir_in + "1998_VR_ke_vi_daily.nc")
```

```python
tm_ke = ds_vr_ke['ke'].mean(["Time", "plevels"])
```

```python
tm_ke_lon = tm_ke.mean("latitude")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import * 

dct_var = tm_ke_lon
# Number of samplepoints
N = dct_var.size
# sample spacing
T = 1 / N
x = np.linspace(0.0, N*T, N+1)
# y = 10 + np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
yf = dct(dct_var.values)
# xf = np.linspace(0.0, 1.0/(2.0*T), int(N/2))
xf = fftfreq(N, T)[:N//2]
yref_synop = xf**(-3)
yref_meso  = xf**(-5/3.)
```

```python
fig, axs = plt.subplots(1,1,figsize=(16,12))

plt.plot(xf, 2.0/N * np.abs(yf[0:int(N/2)]))
plt.plot(xf, yref_synop,color='r')
plt.plot(xf, yref_meso,color='g')
# plt.plot(xf[1:], 2.0/N * np.abs(yf[0:int(N/2)])[1:])
axs.set_yscale('log')
axs.set_xscale('log')
```
